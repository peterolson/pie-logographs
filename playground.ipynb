```python
import pandas as pd

df = pd.read_excel("wiktionary_Pie.xlsx")
df
```

```python
df.rename(columns={"headword": "PIE", "Chinese":"char", "source_url": "references"}, inplace=True)
df
```

```python
df = df.fillna("")

records = df.to_dict(orient="records")
for record in records:
    record["references"] = [record["references"]]
records
```

```python
# create json file
import json
with open("wiktionary_Pie.json", "w", encoding="utf-8") as f:
    json.dump(records, f, ensure_ascii=False, indent=2)
```

```python
# get Dong Chinese data dump
lines = []
with open(r'dictionary_char_2024-02-17.jsonl', encoding='utf-8') as f:
    lines = f.read().splitlines()

line_dicts = [json.loads(line) for line in lines]
dong_df = pd.DataFrame(line_dicts)
dong_df
```

```python
dong_df.columns
```

```python
dong_df = dong_df[dong_df['isVerified'] == True]
reduced = dong_df[['char', 'gloss', 'hint', 'originalMeaning']].fillna("")
# join the gloss and originalMeaning into one column, separated by a semicolon when both are present, otherwise just use the one that is present
reduced['meanings'] = reduced['gloss'] + reduced['originalMeaning'].apply(lambda x: "; " + x if x else "")
reduced = reduced.drop(columns=['gloss', 'originalMeaning'])
reduced = reduced.rename(columns={"hint": "character_hint"})
reduced['references'] = "https://www.dong-chinese.com/wiki/" + reduced['char']
reduced
```

```python
reduced_dict = reduced.to_dict(orient="records")
for record in reduced_dict:
    record["references"] = [record["references"]]
reduced_dict
```

```python
# combine into one list
combined = records + reduced_dict
combined
```

```python
# save to lexicon.json
with open("combined_lexicon.json", "w", encoding="utf-8") as f:
    json.dump(combined, f, ensure_ascii=False, indent=2)
```
